## Lambda Functions <br>
<p>Lambda functions... or "Anonymous Functions" are referring to inline functions with no name. The keyword lambda denotes the no name function, and executes within a single line. Without saving it to a variable; however, it is not able to be used, unless passed in either as a paramater or within list comprehension.<br>Written as "(keyword lambda) (one or more inputs) (colon) (function to be executed)"</p>

#### Syntax

```python
def addTwo(x):   #
    return x + 2

print(addTwo(4))

# Lambda Function Syntax
# Example of calling a lambda function without a variable
print((lambda x: x + 2)(4)) #double parentheses to call the lamda   

# a function into a function
```

#### Saving to a Variable

```python
f_test = lambda x: x + 2

f_test(4)
```

#### Multiple Inputs

```python
# Multiple Inputs with no variable
print((lambda x,y,z: x * y * z)(3,5,8))  #x=3. y=5  z=8

# Multiple Inputs with a variable attached
x_test = lambda x,y,z: x * y * z
print(x_test(3,5,8))   #printing the x_test then calling the assigning the variables their numbers
```

#### Passing a Lambda into a Function

```python
def multiply(f,number):   #making this a function by defining it
    """
        f expects a lambda function
        number expects a number
    """
    return f(number)   
multiply(lambda x: x * x, 4) # lambda 4: 4 * 4   since still inside a comma sets up for the number 4
```

#### Returning a Lambda from a Function

```python
# Regular defined function
def multiply_test(num):
    return num * 4

# Function within a function
def returnFunc():
    test = 4  #showing that you can do work within the function
    
    def multiply(num):
        return num * 2
    return multiply  # only returning this function on line 8
#most likely better to 
f_return = returnFunc()
print(returnFunc())  #this is a funkier version
print(f_return(4))

# Lambda function returned from a regular function
def returnLam(b,c):    #this is the function        the variables are b and c
    return lambda x,a: (x + a) / (b + c) #returning this function  now we are using more variables having X and A
r_lamb = returnLam(2,1)
print(r_lamb(5,5))
```

#### If Statements within Lambdas

```python
# Lambda x : True if (condition) else False
f_condition = lambda num : num * 2 if num > 10 else num + 2  # saving the function to a variable
#             =8 then 12 then 10 
print(f_condition(8))

print(f_condition(12))

print(f_condition(10))
```

## Map <br>
<p>The map function allows you to iterate over an entire list while running a function on each item of the list. This is why the map function works well with lambda's, because it simplifies things and you write less lines of code.<br>The syntax for a map function is "map(function to be used, list to be used)"<br>However, you must be careful, as the map function returns a map object, not a list. To turn it into a list we use the list() type conversion.</p>

#### Syntax

```python
# map(func,iterable(list,dict,tuple,set))   
# Normally the usage of map happens with a pre-defined function -- but we can use lambdas as well
def squared(num,num2):  # mapping over 2 lists
    if num < 10 and num2 < 10:    #giving it a condition
        return num ** 2, num2 ** 2   # we want them turned to the second power
    else:
        return num,num2  # if theyre over 10 it just print the numbers
    
numbers = [4,11,20,3,15,20]  #creating a couple lists for examples   
more_nums = [4,10,3,2,6]  # 2nd list our and our lists are not the same length

squared_nums_map = list(map(squared,numbers,more_nums))  # naming our map then giving it a function
print(squared_nums_map)  # NUMBERS HIGHER THAN 10 ARE RETURNED AS NORMAL WITH ELSE CONDITION ON LINE 6
```

#### Using Lambda's with Map

```python
# map(lambda x: x + 2, list)

# using lambda in map happens in line (or in one line) usually

# list(map(lambda x : x * 2,nums))
                # making a list then wrapping a map
squared_nums_lamb = list(map(lambda x,y : (x ** 2,y ** 2) if x < 10 and y < 10 else (x,y),numbers,more_nums))
print(squared_nums_lamb)

#this is the more efficient way using a map and lambda
```

## Filter() <br>
<p>Filter's are similar to the map function, where you're able to pass a function argument and a list argument and filter out something from the list based on the conditions passed. Similar to the map function, it returns a filter object, so you need to type convert it to a list()</p>

#### Syntax

```python
names = ['Bob','Andy','Max','Evan','Angelica']

def a_names(name):
    if name[0].lower() == 'a':
        return True
    else:
        return False
    
    
new_names_list = list(filter(a_names, names)) #wrapping the list
print(new_names_list) #will only print andy and angelica

print(a_names)
```

#### Using Lambda's with Filter()

```python
new_names_lamb = list(filter(lambda name: True if name[0].lower() == 'a' else False, names))  #prints A names
print(new_names_lamb)
```

## Reduce() <br>
<p>Be very careful when using this function, as of Python 3 it's been moved to the 'functools' library and no longer is a built-in function.<br>The creator of Python himself, says to just use a for loop instead.</p>

#### Syntax

```python
from functools import reduce #could also just use a for loop

# reduce(function,iterable)

list_1 = [2,4,6,8,10] # making a list

def addNums(num1,num2): #num1 refers to
    return num1 + num2

result_add = reduce(addNums,list_1)  #reducing 

print(result_add)

# Subtract a list of numbers
def subtractNums(num1,num2):
    return num1 - num2

result_sub = reduce(subtractNums, list_1)
print(result_sub) #weve been subtracting every number in the list
```

#### Using Lambda's with Reduce()

```python
#  doing everything above with a lambda now 

result_lamb = reduce(lambda x,y: x + y, list_1)
print(result_lamb)
```

## Recursion <br>
<p>Recursion means that a function is calling itself, so it contanstly executes until a base case is reached. It will then push the returning values back up the chain until the function is complete. A prime example of recursion is computing factorials... such that 5! (factorial) is 5*4*3*2*1 which equals 120.</p>

#### Implementing a Base Case

```python
#imagine a russian doll going from bigger to smaller to smaller

def addNums(num): #turning it into its own function 
    # Set base case for recusive function
    if num <= 1:   # if the number is less than or equal to 1 
        print("addNums(1) = 1")  
        return num # and just return the number
    #recursive part of the function
    else:
        print(f"addNums({num}) = {num} + addNums({num - 1})") #formatting string so we can see out values
        return num + addNums(num - 1) #should recursively lower
            # starting with 5 and going down
addNums(5) #running 
```

#### Writing a Factorial Function

```python
# 5 Factorial which is the same as 5 * 4 * 3 * 2 * 1

def factorial(num):  # better visualized on python tutor
    if num <= 1:    # the factorial will be in limbo until at least one is true
        return 1
    else:
        return num * factorial(num - 1)
    
factorial(5)  # 5 then 4 then 3 then 2 would loop until it reaches 1 then they would all execute going back up
```

## Generators <br>
<p>Generators are a type of iterable, like lists or tuples. They do not allow indexing, but they can still be iterated through with for loops. They are created using functions and the yield statement.</p>

#### Yield Keyword <br>
<p>The yield keyword denotes a generator, it doesn't return so it won't leave the function and reset all variables in the function scope, instead it yields the number back to the caller.</p>

```python
def my_range(stop, start, step = 2): #creating a function
    while start < stop:  
        yield start # yield keyword denotes a generator
        start += step
for i in my_range(20, start = 2): # setting my stop at 20 and start at 2
    my_generator_value = i  
    print(my_generator_value)

my_range(20,start = 2)
```
